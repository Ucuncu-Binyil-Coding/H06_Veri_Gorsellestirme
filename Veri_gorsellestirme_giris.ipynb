```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("datalar/flight_delays.csv")
df
```

```python
plt.figure(figsize=(10,5))
plt.title("Aylara göre gecikme süreleri")
sns.barplot(x = df.index, y= df["NK"])
# plt.show()
```

```python
plt.figure(figsize = (14,7))
plt.title("Aylara göre her havalimanın uçuş gecimleri dikak cinsinden")
sns.heatmap(data=df,annot=True)
plt.xlabel("Hava Limanları")
plt.ylabel("Aylar")
plt.show()
```

```python
df = pd.read_csv("datalar/spotify.csv", index_col = 0, parse_dates = True)
df
```

```python
plt.figure(figsize=(5,3))
plt.title("Spotify şarkılarının zamana dayalı dinlenme sayıları")
plt.xlabel("Tarih")
plt.ylabel("Dinlenme")
sns.lineplot(data = df)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.title("Spotify şarkılarının zmana dayalı dinlenme sayıları")
plt.xlabel("Tarih")
plt.ylabel("Dinlenme")
sns.lineplot(data=df["Despacito"],label="Despacito")
sns.lineplot(data=df["Shape of You"],label="Shape of You")
plt.show()
```

```python
df = pd.read_csv("datalar/iris.csv")
df
```

```python
plt.figure(figsize=(5,3))
plt.title("Iris dataset falan")
plt.xlabel("Uzunluk")
plt.ylabel("Çiçek sayısı")
sns.histplot(data=df["petal_length"],kde=True)
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.title("Iris dataset falan")
plt.xlabel("Uzunluk")
plt.ylabel("Çiçek sayısı")
sns.kdeplot(data=df["petal_length"],fill=True)
sns.kdeplot(data=df["petal_width"],fill=True)
plt.show()
```

```python
sns.jointplot(x=df["petal_length"],y= df["sepal_width"], kind="kde")
plt.show()
```

```python
sns.histplot(data=df,x= df["petal_length"], hue= "species")
plt.show()
```

```python
plt.figure(figsize=(5,3))
plt.title("Iris dataset falan")
plt.xlabel("petal_length")
# plt.ylabel("Çiçek sayısı")
sns.kdeplot(x = df["petal_length"], hue = df["species"], fill=True)
plt.show()
```

```python
df = pd.read_csv("datalar/insurance.csv")
df
```

```python
plt.figure(figsize=(10,8))
plt.title("Vücut kitle endeksine göre yıllık sigorta ödemeleri")  
plt.xlabel("Vücut kitle endeksi")
plt.ylabel("Yıllık ödeme")
sns.scatterplot(x = df['bmi'], y= df['charges'])
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title("Vücut kitle endeksine göre yıllık sigorta ödemeleri")  
plt.xlabel("Vücut kitle endeksi")
plt.ylabel("Yıllık ödeme")
sns.regplot(x = df['bmi'], y= df['charges'])
# plt.hlines([10000],df['bmi'].min(),df['bmi'].max())
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title("Vücut kitle endeksine göre yıllık sigorta ödemeleri")  
plt.xlabel("Vücut kitle endeksi")
plt.ylabel("Yıllık ödeme")
sns.scatterplot(x = df['bmi'], y= df['charges'],hue=df['smoker'])
plt.show()
```

```python
sns.lmplot(data=df,x = 'bmi', y= 'charges',hue='smoker')
plt.show()
```

```python
df = pd.read_csv("datalar/SampleSuperstore.csv")
df
```

```python
numeric_cols = ["Sales","Quantity","Discount","Profit"]
fig,axes = plt.subplots(1,len(numeric_cols),figsize=(10,6))
plt.subplots_adjust(wspace=0.5)
# df[numerical]
for col,ax in zip(numeric_cols,axes):
    ax.set_xlabel(col)
    ax.set_ylabel("")
    ax.set_xticklabels([])
    sns.boxplot(df[col],ax=ax)
```
